```python
def bdate_year_parser(bday):
    if pd.isna(bday):
        return None
    bday = bday.split(".")
    if len(bday) == 3:
        return pd.to_datetime(".".join(bday), format="%d.%m.%Y")
    else:
        return None
df["bdate"].apply(bdate_year_parser)
```

```python
df["bdate"] = df["bdate"].apply(bdate_year_parser)
```
